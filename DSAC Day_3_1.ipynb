### Clustering Practice

```python
from scipy.cluster.hierarchy import dendrogram, linkage

#link_dist = linkage(power1, metric = 'euclidean', method = 'ward')
#dendrogram(link_dist, labels = power1.index)

link_dist = linkage(power2, metric = 'euclidean', method = 'ward')
dendrogram(link_dist, labels = power2.index)

# link_dist
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

link_dist = linkage(power2, metric = 'euclidean', method = 'centroid')      # ward, centroid, minimum, maximum, median
dendrogram(link_dist, labels = power2.index)
```

### Clustering by KMeas()

```python
from sklearn.cluster import KMeans
k = KMeans(n_clusters = 4)
k.fit(power2)
```

```python
print(k.labels_)
# print(k.cluster_centers_)
```

```python
power2['클러스터종류']=k.labels_
power2['클러스터종류']
```

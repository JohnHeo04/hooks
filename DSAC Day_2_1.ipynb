```python
# example coding: calculating pi
import matplotlib.pyplot as plt
%matplotlib inline  

N=10000
x=np.random.rand(N)
y=np.random.rand(N)

cnt=0
for n in range(N):
    if ((x[n]*x[n]) +  (y[n]*y[n])) < 1.0:
        cnt = cnt + 1
    else :
        continue
pi = 4*cnt/N
print(pi)

plt.figure(figsize=(5,5))    # plotting a circle
plt.scatter(x,y,s=1)         # s: marker size

a = np.linspace(0,1,100)
b = (1-a*a)**0.5
plt.plot(a,b,c='r')
```
